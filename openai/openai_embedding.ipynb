```python
from openai import OpenAI
client = OpenAI(api_key="YOUR_API_KEY")

response = client.embeddings.create(
    input="왕",
    model="text-embedding-3-small"
)

print(response.data[0].embedding)
```

```python
def embedding(input):
    response = client.embeddings.create(
        input=input,
        model="text-embedding-3-small"
    )

    return response.data[0].embedding
```

```python
king = embedding('왕')
queen = embedding('여왕')
man = embedding('남자')
woman = embedding('여자')
```
