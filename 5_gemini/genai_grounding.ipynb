```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
```

```python
from google import genai
from google.genai.types import Tool, GenerateContentConfig, GoogleSearch

client = genai.Client(api_key=GOOGLE_API_KEY)
model_id = "gemini-2.0-flash"

google_search_tool = Tool(
    google_search = GoogleSearch()
)

response = client.models.generate_content(
    model=model_id,
    contents="현재 비트코인 가격을 말하고 상승할지 하락할지 예측하고, 그 근거를 20 단어 이내로 설명하세요.",
    config=GenerateContentConfig(
        tools=[google_search_tool],
    )
)

for each in response.candidates[0].content.parts:
    print(each.text)
```

```python
from google.genai import types

response = client.models.generate_content(
    model='gemini-2.0-flash',
    contents="현재 비트코인 가격을 말하고 상승할지 하락할지 예측하고, 그 근거를 20 단어 이내로 설명하세요.",
    config=GenerateContentConfig(
        tools=[
            Tool(
                google_search=types.GoogleSearchRetrieval(
                dynamic_retrieval_config=types.DynamicRetrievalConfig(
                dynamic_threshold=0.6))
            )
        ]
    )
)
for each in response.candidates[0].content.parts:
    print(each.text)
print(response.candidates[0].grounding_metadata)
```

```python
from google.genai import types
from pprint import pprint

response = client.models.generate_content(
    model='gemini-2.0-flash',
    contents="현재 비트코인 가격을 말하고 상승할지 하락할지 예측하고, 그 근거를 20 단어 이내로 설명하세요.",
    config=GenerateContentConfig(
        tools=[
            Tool(
                google_search=types.GoogleSearchRetrieval(
                dynamic_retrieval_config=types.DynamicRetrievalConfig(
                dynamic_threshold=0.6))
            )
        ]
    )
)

# 응답 텍스트 출력
print("=== 생성된 응답 ===")
for each in response.candidates[0].content.parts:
    print(each.text)
print("\n")

# 근거와 확률 쌍 출력
print("=== 근거와 확률 쌍 ===")

# 그라운딩 메타데이터 가져오기
metadata = response.candidates[0].grounding_metadata
chunks = metadata.grounding_chunks if hasattr(metadata, 'grounding_chunks') else []

# 각 지원 세그먼트(문장)에 대해 처리
for i, support in enumerate(metadata.grounding_supports):
    segment_text = support.segment.text
    print(f"\n[문장 {i+1}] {segment_text}")

    # 각 소스와 해당 확률 점수 출력
    for j, (chunk_idx, score) in enumerate(zip(support.grounding_chunk_indices, support.confidence_scores)):
        if chunk_idx < len(chunks) and hasattr(chunks[chunk_idx], 'web'):
            source = chunks[chunk_idx].web.title
            print(f"  - 근거 {j+1}: {source} (신뢰도: {score:.4f})")
        else:
            print(f"  - 근거 {j+1}: 소스 #{chunk_idx} (신뢰도: {score:.4f})")

# 검색 쿼리 출력
print("\n=== 검색 쿼리 ===")
for i, query in enumerate(metadata.web_search_queries):
    print(f"쿼리 {i+1}: {query}")
```
